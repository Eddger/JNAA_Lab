```python
# Timeseries plot
# Plot the rainfall timeseries for January and Feburary

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(data[:, 0], data[:, 1], color="green", lw=1.5, ls="--", marker="*", label="Jan") #We specify some extra attributes here
ax.plot(data[:, 0], data[:, 2], color="red", lw=1.5, ls="--", marker="*", label="Feb")

# axis labelling
ax.set_ylabel("Rainfall Values [mm]", fontsize=15)
ax.set_xlabel("Year", fontsize=15)
ax.set_title(
    "Rainfall for 1983 t0 2017 ", fontsize=15
)  # add title
ax.legend()  # add legend
fig.savefig("1983_2017_rain.pdf")  # save figure to pdf
```
