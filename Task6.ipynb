```python
import numpy as np
import matplotlib as mlp
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#reading file
a=pd.read_excel('C:/Users/USER/Desktop/OneQuantum-Ghana-main/GAUGE_PRECIP/WEATHER CALCULATOR.xlsx')
```

```python
#creating a new column with heading Time
a['Time']=['12','15','18','21','00','03','06','09','012','NaN']
```

```python
#ac=['12','15','18','21','00','03','06','09','012']
```

```python
#Plotting graphs 
fig, axes = plt.subplots(figsize=(10,6))

ax1=axes
ax1.plot(a.Time,a.TEMP,color='red',lw=2, marker='*', markersize=8)
ax1.set_xlabel('TIME [UTC]', fontsize=20)
ax1.set_ylabel('TEMPERATURE [$^o$ C]', fontsize=20)

axes.set_title('TEMPERATURE AND RELATIVE HUMIDITY', fontsize=20)

#plotting both columns with same x-axis 
ax2=axes.twinx()
ax2.plot(a.Time,a.RH,color='blue',lw=2, marker='o', markersize=8)
ax2.set_xlabel('Time UTC',fontsize=20)
ax2.set_ylabel('RELATIVE HUMIDITY [%]', fontsize=20)
```
