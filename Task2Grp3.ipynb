```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rain_data =np.genfromtxt('Tamale_CRU_rainfall.csv',delimiter=',')
rain_data
```

```python
years=rain_data[1:,0]
aug=rain_data[1:,8]
jun=rain_data[1:,6]
total=np.sum(rain_data[1:,1:], axis=1)
total_mean=np.mean(rain_data[1:,1:], axis=1)
```

```python
fig, axes=plt.subplots(ncols=2, figsize=(10,6))

ax1=axes[0]
ax1.bar(years, aug)
ax1.set_xlabel('Year',fontsize=20)
ax1.set_ylabel('Aug. Rainfall', fontsize=20)
ax1.set_title('August Annual Rainfall', fontsize=20)

ax2=axes[1]
ax2.plot(years, total, color="green", lw=1.5, ls="--", marker="*", label="Jan")
ax2.set_xlabel('Years', fontsize=20)
ax2.set_ylabel('Total Rainfall [mm]', fontsize=15)
ax2.set_title('Annual Rainfall Total',fontsize=20)
ax2.legend()
```

```python
fig, axes=plt.subplots(figsize=(10,6))

ax1=axes
ax1.bar(years, jun)
ax1.set_xlabel('Year', fontsize=20)
ax1.set_ylabel('June Rainfall', fontsize=20)

ax2=axes.twinx()
ax2.plot(years, total_mean, color="green", lw=1.5, ls="--", marker="*", label="Jan")
ax2.set_xlabel('Years', fontsize=20)
ax2.set_ylabel('Mean Rainfall [mm]', fontsize=20)
ax2.set_title('Annual Mean Rainfall',fontsize=30)
ax2.legend()
```
