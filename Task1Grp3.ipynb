Read the rr_temp_rh_24_30May21.csv. The file contains gridded data, the columns describe lat: latitude and lon: longitude for each datapoint, rr: total rainfall, tn weekly-mean daily-minimum temperature, tx: weekly-mean daily-maximum temperature and rh mean relative humidity

Generate spatial map for rainfall, maximum and minimum temperature and relative humidity
Save figure to pdf and png
Interpret the figures

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cf
import cartopy.io.shapereader as shapereader

from pathlib import Path
```

```python
dtxt=np.genfromtxt("C:/Users/USER/Desktop/OneQuantum-Ghana-main/data/rr_temp_rh_24_30May21.csv", delimiter=",", skip_header=1)

lat=dtxt[:,0]
lon=dtxt[:,1]
rr=dtxt[:,2]
tx=dtxt[:,3]
tn=dtxt[:,4]
rh=dtxt[:,5]
```

```python
shp_path = ('C:/Users/USER/Desktop/OneQuantum-Ghana-main/data/gh_shp/gh_wgs16dregions.shp')
adm1_shapes=list(shapereader.Reader(str(shp_path)).geometries())
```

```python
fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection=ccrs.PlateCarree())

ax.set_extent([-4, 2, 4, 12], crs=ccrs.PlateCarree())

ax.add_geometries(adm1_shapes, ccrs.PlateCarree(), edgecolor='black', facecolor='none', alpha=0.8)

ax.add_feature(cf.BORDERS)
ax.add_feature(cartopy.feature.COASTLINE)
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
#ax.add_feature(cf.LAKES, edgecolor='black')


ax.set(title='Total Rainfall - 30-May-21')
splot = ax.scatter(lon,lat,25,c=rr, cmap=cm.get_cmap('Blues'), alpha=1)
fig.colorbar(splot,label="Rainfall Amount (mm)", orientation="vertical", extend='both')
#fig.savefig('Rainfall.pdf')
#fig.savefig('Rainfall.png')
```
